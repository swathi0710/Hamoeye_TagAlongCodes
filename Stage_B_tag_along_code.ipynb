Q12:r square value

```python
from sklearn.model_selection import train_test_split

sample = data[["T2","T6"]]

x=sample["T2"].values.reshape(-1,1)
y=sample["T6"].values.reshape(-1,1)

x1, x2, y1, y2 = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

m=LinearRegression()
m.fit(x1,y1)

predicted=m.predict(x2)
```

```python
from sklearn.metrics import r2_score

r2_score(y2,predicted)
```

Q13: Normalise and MAE

```python
from sklearn.preprocessing import MinMaxScaler

data=data.drop(["date","lights"],axis=1)
s=MinMaxScaler()
s_data=pd.DataFrame(s.fit_transform(data),columns=data.columns)
```

```python
Y=s_data["Appliances"]
X=s_data.drop(["Appliances"],axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

lr=LinearRegression()
lr.fit(X_train,Y_train)

pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae=mean_absolute_error(Y_test,pred)
mae
```

Q14: Residual sum of squares

```python
rss=np.sum(np.square(Y_test-pred))
rss
```

Q15:RMSE

```python
l_rmse =np.sqrt(mean_squared_error(Y_test, pred))
l_rmse
```

```python
from sklearn.linear_model import Ridge

rr=Ridge(alpha=0.4)
rr.fit(X_train, Y_train)
r_pred=rr.predict(X_test)
```

Q18


```python
r_rmse=np.sqrt(mean_squared_error(Y_test, r_pred))
r_rmse
```
